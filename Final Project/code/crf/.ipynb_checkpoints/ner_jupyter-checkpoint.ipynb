```python
file2 = 'unlabeled_train'
with open(file2) as f:
    raw_data = f.readlines()
```

```python
def prep(word):
    word = word.strip()
    word = word.replace('[', '')
    word = word.replace(']', '')
    return word

raw_data = [prep(word) for word in raw_data]
```

```python
raw_data
```

```python
data = []
for word in raw_data:
    if word != '' and '.pos' not in word:
        data.append(word)
```

```python
sent_set = []
sent = []
for line in data:
    if line[0:2] == '==':
        if not sent:
            pass
        else:
            sent_set.append(sent)
            sent = []
    else:
        for word in line.split():
            sent.append(tuple(word.split('/')))
            
```

```python
sent_set
```

```python
save_file = 'unlabeled_set'
with open(save_file, 'w') as f:
    for sent in sent_set:
        f.write('%s\n' % sent)
```
